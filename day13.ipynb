```python
def lprint(l):
    for row in l:
        print(row)
    return
```

```python
def pprint(carts, track):
    track = [list(row) for row in track]
    for cart in carts:
        current = track[cart.y][cart.x]
        if current in '<>^v':
            track[cart.y][cart.x] = 'X'
        else:
            track[cart.y][cart.x] = cart.dir
    track = [''.join(row) for row in track]
    lprint(track)
    return
```

```python
from dataclasses import dataclass
from functools import total_ordering

@dataclass
class Cart:
    pos: tuple
    dir: str
    state: str
        
    @property
    def x(self):
        return self.pos[0]
    
    @property
    def y(self):
        return self.pos[1]
    
    @total_ordering
    def __eq__(self, other):
        return self.pos == other.pos
    
    @total_ordering
    def __lt__(self, other):
        return (self.y, self.x) < (other.y, other.x)
    
    def shift(self):
        moves = {
            '>': (1,0),
            '<': (-1, 0),
            'v': (0, 1),
            '^': (0, -1)
        }
        m = moves[self.dir]
        x = self.x + m[0]
        y = self.y + m[1]
        self.pos = (x, y)
        return self
        
    def swivel(self):
        """State is memory of last move, not knowledge of next move"""
        turns = "LSRL"
        self.state = turns[turns.find(self.state) + 1]

        lookup = {'L': '<v>^<', 'R': '<^>v<', 'S': '<<>>^^vv'}
        s = lookup[self.state]
        self.dir = s[s.find(self.dir) + 1]
        
        return self

    def move(self, track):  
        # Move then change dir.
        self = self.shift()
        
        # Change dir.
        turns = {'\\': {'>': 'v', '<': '^', 'v': '>', '^': '<'},
                 '/': {'>': '^', '<': 'v', 'v': '<', '^': '>'},
                 '-': {'>': '>', '<': '<'},
                 '|': {'v': 'v', '^': '^'},
                 }
        t = track[self.y][self.x]
        if t == '+':
            self = self.swivel()
        else:
            self.dir = turns[t][self.dir]

        return self
```
